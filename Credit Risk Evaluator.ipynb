```python
# Converts categorical data to numeric
le = LabelEncoder()
train_df['application_type'] = le.fit_transform(train_df.application_type.values)
test_df['application_type'] = le.fit_transform(test_df.application_type.values)
```

```python
# Converts categorical data to numeric
le = LabelEncoder()
train_df['debt_settlement_flag'] = le.fit_transform(train_df.debt_settlement_flag.values)
test_df['debt_settlement_flag'] = le.fit_transform(test_df.debt_settlement_flag.values)
```

```python
# Converts categorical data to numeric

le = LabelEncoder()
train_df['hardship_flag'] = le.fit_transform(train_df.hardship_flag.values)
test_df['hardship_flag'] = le.fit_transform(test_df.hardship_flag.values)
```

```python
# Converts categorical data to numeric

le = LabelEncoder()
train_df['home_ownership'] = le.fit_transform(train_df.home_ownership.values)
test_df['home_ownership'] = le.fit_transform(test_df.home_ownership.values)
```

```python
# Converts categorical data to numeric

le = LabelEncoder()
train_df['initial_list_status'] = le.fit_transform(train_df.initial_list_status.values)
test_df['initial_list_status'] = le.fit_transform(test_df.initial_list_status.values)
```

```python
# Converts categorical data to numeric

le = LabelEncoder()
train_df['pymnt_plan'] = le.fit_transform(train_df.pymnt_plan.values)
test_df['pymnt_plan'] = le.fit_transform(test_df.pymnt_plan.values)
```

```python
# Converts categorical data to numeric

le = LabelEncoder()
train_df['target'] = le.fit_transform(train_df.target.values)
test_df['target'] = le.fit_transform(test_df.target.values)
```

```python
# Converts categorical data to numeric

le = LabelEncoder()
train_df['verification_status'] = le.fit_transform(train_df.verification_status.values)
test_df['verification_status'] = le.fit_transform(test_df.verification_status.values)
```

```python
# add missing dummy variables to testing set

for col in train_df.columns:
    if col not in test_df.columns:
        test_df[col]=0
```

```python
# Separates target feature for testing data

y_train = train_df["target"]
X_train = train_df.drop(columns = ["target"])
X_train.head()
```

```python
# Separates target feature for testing data

y_test = test_df["target"]
X_test = test_df.drop(columns = ["target"])
X_test.head()
```

```python
# Splits the training data

print(X_train.shape, y_train.shape)
```

```python
# Splits the training data

print(X_test.shape, y_test.shape)
```

```python
# Trains the Logistic Regression model on the unscaled data and print the model score

logisticRegr = LogisticRegression(
    solver='lbfgs', 
    max_iter=100,
    random_state=0
)
logisticRegr.fit(X_train, y_train)
print("LogisticRegression score: ", logisticRegr.score(X_test, y_test))
```

```python
# Trains a Random Forest Classifier model and print the model score
randomForestClass = RandomForestClassifier(random_state=0)
randomForestClass.fit(X_train, y_train)
print("RandomForestClassifier score: ", randomForestClass.score(X_test, y_test))
```

```python
# Scales the data

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Trains the Logistic Regression model on the scaled data and print the model score

logisticRegr_scaled = LogisticRegression(
    solver='lbfgs', 
    max_iter=100,
    random_state=0
)
logisticRegr_scaled.fit(X_train_scaled, y_train)
print("LogisticRegression scaled score: ", logisticRegr_scaled.score(X_test_scaled, y_test))
```

```python
# Trains a Random Forest Classifier model on the scaled data and print the model score

randomForestClass_scaled = RandomForestClassifier(random_state=0)
randomForestClass_scaled.fit(X_train_scaled, y_train)
print("RandomForestClassifier scaled score: ", randomForestClass_scaled.score(X_test_scaled, y_test))
```
